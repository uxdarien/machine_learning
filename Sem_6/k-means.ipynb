Módulo kmeans de sklearn [aquí](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans


X, y = make_blobs(n_samples=100, centers=3, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)

def update_plot(frame):
    kmeans.partial_fit(X[frame])
    if frame % 10 == 0:
        plt.cla()
        plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
        plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red')
        plt.title(f"Iteration {frame}")


fig = plt.figure()
ani = animation.FuncAnimation(fig, update_plot, frames=len(X), interval=200)

plt.show()
```
