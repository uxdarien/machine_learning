# **Support Vector Boundary**




Información sobre base de datos isotrópica [aquí](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html#sklearn.datasets.make_blobs) \\
Información sobre support vector classifier [aquí](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC)

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs # Libreria para crear datos sintéticos
```

```
# we create 40 separable points
X, y = make_blobs(n_samples=150, centers=2, random_state=6)
_, ax = plt.subplots(figsize=(6,4))
ax.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap="Paired")
```

```
clf = svm.SVC(kernel='linear', C=1000)
clf.fit( X , y )
```

```
## Plot decision boundaries
prec = 0.1
# Valores minimos y maximos para crear el grid de las fronteras de decision
x_min, x_max = X[:, 0].min()-prec, X[:, 0].max()+prec
y_min, y_max = X[:, 1].min()-prec, X[:, 1].max()+prec

# Crear el grid de las fronteras de decision
# x_grid, y_grid = np.meshgrid(np.arange(x_min, x_max, prec), np.arange(y_min, y_max, prec)) #Option 1
x_grid, y_grid = np.meshgrid(np.linspace(x_min,x_max,300), np.linspace(y_min,y_max,300)) #Option 2
```

```
# Predecir los valores del grid
predictions = clf.decision_function(np.c_[x_grid.ravel(), y_grid.ravel()])

predictions = clf.predict(np.c_[x_grid.ravel(), y_grid.ravel()])

print (predictions.shape)

# Encajar las predicciones al tamaño del grid
predictions = predictions.reshape(x_grid.shape)
print (predictions.shape)
```

```
_, ax = plt.subplots (figsize=(8,5))
# ax.scatter( _ , _ , c = _ , s = 30, cmap = plt.cm.Paired)

# plot decision boundary and margins
ax.contour( x_grid , y_grid , predictions , colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

ax.pcolormesh(x_grid, y_grid, predictions, cmap="Paired")


# plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
ax.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='Paired')


ax.set_xlim(x_grid.min()-.1, x_grid.max()+.1)
ax.set_ylim(y_grid.min()-.1, y_grid.max()+.1)

ax.set_xticks(())
ax.set_yticks(())
```
